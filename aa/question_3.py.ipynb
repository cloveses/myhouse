```python
import requests
import re
import random
import time

def param_html(html):
    datas = []
    res = re.findall(r'<div class="houseInfo">(.*?)</div>',html,re.S|re.M)
    for r in res:
        room = []
        region = re.findall(r'>(.*?)</a>', r, re.S|re.M)[0].strip()
        room.append(region)
        room.extend(re.findall(r'</span>(.*?)<span',r, re.S|re.M))
        datas.append(room)
    return datas

def get_float(data):
    data = data[2][:-2]
    try:
        area = float(data)
    except:
        area = 0
    return area

def method_1():
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
        'Accept':'text/html;q=0.9,*/*;q=0.8',
        'Accept-Charset':'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
        'Accept-Encoding':'gzip',
        'Connection':'close',
        'Referer':'http://www.baidu.com/link?url=_andhfsjjjKRgEWkj7i9cFmYYGsisrnm2A-TN3XZDQXxvGsM9k9ZZSnikW2Yds4s&wd=&eqid=c3435a7d00006bd600000003582bfd1f'
        }
    sess = requests.Session()
    url = 'https://bj.lianjia.com/ershoufang/pg{}/'
    datas = []
    for i in range(1,4):
        r = sess.get(url.format(i),headers=headers)
        if r.ok:
            datas.extend(param_html(r.text))
        time.sleep(random.randint(3,8))
    
    datas.sort(key=get_float)
    return(datas)
        
print(method_1())
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs

def method_2():
    url = 'https://bj.lianjia.com/ershoufang/pg{}/'
    datas = []
    br = webdriver.Firefox()
    datas = []
    for i in range(1,11):
        br.get(url.format(i))
        content = br.page_source.encode('utf-8')
        soup = bs(content, 'lxml')
        for s in soup.find_all('div','address'):
            td = s.text.split('/')
            td = [t.strip() for t in td]
            datas.append(td)
        time.sleep(random.randint(3,8))
    datas.sort(key=get_float)
    br.close()
    return datas

print(method_2())
    
```
